```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import files

```python
# import inventory file
inventory = pd.read_csv('library_inventory_matches.csv')
inventory.head()
```

```python
# import dictionary
data_dict = pd.read_csv('library_System_Data_Dictionary.csv')
data_dict.head()
```

```python
# import checkout files from 2014-2016
checkout2014 = pd.read_csv('librarycheckouts_2014.csv')
checkout2014.head()
```

```python
checkout2015 = pd.read_csv('librarycheckouts_2015.csv')
checkout2015.head()
```

```python
checkout2016 = pd.read_csv('librarycheckouts_2016.csv')
checkout2016.head()
```

# Merge

```python
# concatenating all checkouts into one dataframe
concat1 = pd.concat([checkout2014, checkout2015])
concat1.head()
```

```python
checkout_full = pd.concat([concat1, checkout2016])
checkout_full.head()
```

```python
inventory.head()
```

```python
# checkouts merged with inventory to add authors to checkout details
full = pd.merge(checkout_full, inventory, how='left', on='BibNumber')
full.head()
```

```python
# delete additional unnamed columns and duplicated columns
full2 = full.drop(columns=['Unnamed: 0_y', 'Unnamed: 0_x', 'ItemType_y'])
full2.head()
```

```python
# rename column to enable merge
full3 = full2.rename(columns = {'ItemType_x': 'Code'})
full3.head()
```

```python
# merge full with dictionary to link media types
complete = pd.merge(full3, data_dict, how='left', on='Code')
complete.head()
```

# Drill down to books

```python
# new table with just the information needed
breakdown = complete[['CheckoutDateTime', 'Title', 'Author', 'Description', 'Format Group', 'Format Subgroup']]
breakdown.head()
```

```python
# narrow down dataframe to only books
books = breakdown[breakdown['Format Subgroup'] == 'Book']
books.head()
```

```python
books.to_csv('books.csv')
```
