```python
words = word_tokenize(content1)
print(f"The total number of words in the text is {len(words)}")
```

```python
# find the frequency of words
fdist = FreqDist(words)

# print 10 most common
fdist.most_common(10)
```

```python
# create empty list to store words
words_no_punc = []

#iterate through the words list to remove punctuation
for word in words:
    if word.isalpha():
        words_no_punc.append(word.lower())
        
# print number of words without punctuation
print(f"The total number of words without punctualtion is {len(words_no_punc)}")
```

```python
fdist = FreqDist(words_no_punc)
fdist.most_common(10)
```

```python
# Download and import list of stopwords
nltk.download("stopwords")
from nltk.corpus import stopwords

# list of stopwords
stopwords_list = stopwords.words("english")
print(stopwords_list)
```

```python
# create and empty list to store clean words
clean_words = []

# Iterate through the words_no_punc list and add non stopwords to the new clean_words list
for word in words_no_punc:
    if word not in stopwords_list:
        clean_words.append(word)
        
print(f"The total number of words without punctuation and stopwords is {len(clean_words)}")
```

```python
fdist = FreqDist(clean_words)
fdist.most_common(10)
```

```python
# Update the stopwords list
stopwords_list.extend(["said", "one", "like", "come", "would", "could", "us", "go", "see", "still"])

# create an empty list to store clean words
clean_words1 = []

# Iterate through the words_no_punc list and add non stopwords to the new clean_words list
for word in words_no_punc:
    if word not in stopwords_list:
        clean_words1.append(word)
```

```python
fdist = FreqDist(clean_words1)
fdist.most_common(10)
```

```python
# Convert word list to a single string
clean_words_string = " ".join(clean_words1)

wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='magma')
wordcloud.generate(clean_words_string)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
