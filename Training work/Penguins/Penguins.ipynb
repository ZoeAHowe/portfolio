```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import files

```python
penguin = pd.read_csv('data/penguins.csv')
```

```python
penguin.head()
```

```python
# find out shape of df
penguin.shape
```

```python
# find out info of columns
penguin.info()
```

```python
# find if there are any nulls
penguin.isnull().sum()
```

```python
# what islands are in the columns
penguin['island'].unique()
```

```python
# what sexes are in the columns
penguin['sex'].unique()
```

# Visualisations - Pandas

```python
species = penguin.species.value_counts()
species.plot(kind='pie', autopct='%.2f%%')
```

```python
island = penguin.island.value_counts()
island.plot(kind='pie', autopct='%.2f%%')
```

```python
sex = penguin.sex.value_counts()
sex.plot(kind='pie', autopct='%.2f%%')
```

# Remove Unnamed column

```python
penguin = penguin.drop('Unnamed: 0', axis=1)
```

# Group Penguins by species

```python
penguin.groupby('species').describe()
```

```python
adelie = penguin[penguin['species']=='Adelie']
adelie.head()
```

```python
gentoo = penguin[penguin['species'] == 'Gentoo']
gentoo.head()
```

```python
chinstrap = penguin[penguin['species'] == 'Chinstrap']
chinstrap.head()
```

# Visualisations - Seaborn

```python
pairplot_figure = sns.pairplot(penguin, hue='species')
pairplot_figure.fig.set_size_inches(9, 6.5)
```

# Scatter plot

```python
# Body mass vs Flipper Length
sns.scatterplot(data=penguin, x='flipper_length_mm', y='body_mass_g', hue='species')
```

```python
# Bill length vs Bill depth
sns.scatterplot(data=penguin, x='bill_length_mm', y='bill_depth_mm', hue='species')
```

# Histogram

```python
# Flipper length
sns.histplot(x='flipper_length_mm', data=penguin)
plt.title('Flipper Length');
```

```python
sns.histplot(y='flipper_length_mm', data=penguin) # change Flipper Length to y axis to show on the side
plt.title('Flipper Length');
```

```python
sns.histplot(x='flipper_length_mm', data=penguin, binwidth=3) # binwidth with change width of columns on histogram
plt.title('Flipper Length');
```

```python
sns.histplot(x='flipper_length_mm', data=penguin, kde=True) # kde will add distribution curve
plt.title('Flipper Length');
```

```python
sns.histplot(x='flipper_length_mm', data=penguin, hue='species')
plt.title('Flipper Length');
```

# Bar plot

```python
sns.barplot(x='species', y='flipper_length_mm', data=penguin)
plt.xlabel('Species')
plt.ylabel('Flipper Length (mm)')
plt.title('Flipper Length by Species');
```

```python
sns.barplot(x='species', y='flipper_length_mm', data=penguin, hue='sex')
plt.xlabel('Species')
plt.ylabel('Flipper Length (mm)')
plt.title('Flipper Length for Penguin Species by Sex');
```

# Box plot

```python
sns.boxplot(x='species', y='flipper_length_mm', data=penguin)
plt.title('Flipper Length for Penguin Species');
```

```python
sns.boxplot(x='species', y='flipper_length_mm', data=penguin, hue='sex')
plt.xlabel('Species')
plt.ylabel('Flipper Length (mm)')
plt.title('Flipper Length for Penguin Species by Sex');
```

# Violin plot

```python
sns.violinplot(x='species', y='flipper_length_mm', data=penguin)
plt.xlabel('Species')
plt.ylabel('Flipper Length (mm)')
plt.title('Flipper Length for Penguin Species');
```

```python
sns.violinplot(x='species', y='flipper_length_mm', data=penguin, hue='sex')
plt.xlabel('Species')
plt.ylabel('Flipper Length (mm)')
plt.title('Flipper Length for Penguin Species by Sex');
```

# Facet grid

```python
sns.FacetGrid(penguin, col='island', row='sex').map(sns.histplot, 'flipper_length_mm');
```

```python
sns.FacetGrid(penguin, col='island', row='sex').map(sns.distplot, 'flipper_length_mm');
```
