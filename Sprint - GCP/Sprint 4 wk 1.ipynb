```python
import pandas as pd
import hashlib
import uuid
```

```python
df = pd.read_csv('source_data.csv')
df.head(5)
```

## Hashed Social Security Number

```python
df['social'] = df['social'].astype(str)

df['hashed_social'] = df['social'].apply(lambda ssn: hashlib.sha256(ssn.encode()).hexdigest() if ssn != "nan" else "NaN")
```

```python
df.head(5)
```

## Mask phone number

```python
df['pseudo_phone'] = df['phone'].apply(lambda x: 'XXX-XXX-' + x[-4:])
```

```python
df.head(5)
```

```python
df['UniqueID'] = [uuid.uuid4() for _ in range(len(df))]
```

```python
df.columns
```
